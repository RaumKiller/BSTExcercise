# Binary Search Tree Exercise
Modifications of the BST program to include node counters, repeated value handling, and traversal adjustments. Also includes $T(n)$ asymptotic complexity analysis.

```python
from collections import deque

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.count = 1 # counter for repeated values

def insert(node, key):
    if node is None:
        return Node(key)
    
    if key == node.key:
        node.count += 1
        return node
    elif key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)
    return node

def search(root, key):
    if root is None or root.key == key:
        return root
    if root.key < key:
        return search(root.right, key)
    return search(root.left, key)

def inorder(root):
    if root is not None:
        inorder(root.left)
        for _ in range(root.count):
            print(root.key, end=" ")
        inorder(root.right)

def preorder(root):
    if root is not None:
        for _ in range(root.count):
            print(root.key, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root is not None:
        postorder(root.left)
        postorder(root.right)
        for _ in range(root.count):
            print(root.key, end=" ")

def height(node):
    if node is None:
        return 0
    else:
        left_depth = height(node.left)
        right_depth = height(node.right)
        if left_depth > right_depth:
            return (left_depth + 1)
        else:
            return (right_depth + 1)

def print_specific_level(root, level):
    if root is None:
        return
    if level == 1:
        for _ in range(root.count):
            print(root.key, end=" ")
    elif level > 1:
        print_specific_level(root.left, level - 1)
        print_specific_level(root.right, level - 1)

def print_level_order(root):
    h = height(root)
    for i in range(1, h+1):
        print_specific_level(root, i)
        print()

def print_leaf_nodes(root):
    if not root:
        return
    if not root.left and not root.right:
        for _ in range(root.count):
            print(root.key, end=" ")
    if root.left:
        print_leaf_nodes(root.left)
    if root.right:
        print_leaf_nodes(root.right)

def print_non_leaf_node(root):
    if root is None or (root.left is None and root.right is None):
        return
    if root.left is not None or root.right is not None:
        for _ in range(root.count):
            print(root.key, end=" ")
    print_non_leaf_node(root.left)
    print_non_leaf_node(root.right)

def max_value_node(node):
    current = node
    while current and current.right is not None:
        current = current.right
    return current

def delete_node(root, key):
    if root is None:
        return root
 
    if key < root.key:
        root.left = delete_node(root.left, key)
    elif key > root.key:
        root.right = delete_node(root.right, key)
    else:
        if root.count > 1:
            root.count -= 1
            return root
            
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
 
        temp = max_value_node(root.left)
        root.key = temp.key
        root.count = temp.count
 
        temp.count = 1
        root.left = delete_node(root.left, temp.key)
 
    return root
```

```python
if __name__ == '__main__':
    root = None

    values = [50, 30, 20, 40, 70, 60, 80, 10, 50, 30, 70, 70]
    print("Inserting values:", values)
    for v in values:
        root = insert(root, v)

    print("\nInorder Traversal:")
    inorder(root)
    print("\n\nPreorder Traversal:")
    preorder(root)
    print("\n\nPostorder Traversal:")
    postorder(root)
    print("\n\nLevel Order Traversal:")
    print_level_order(root)

    print("Leaf Nodes:")
    print_leaf_nodes(root)
    print("\n\nNon-Leaf Nodes:")
    print_non_leaf_node(root)
    print("\n\n---")
    
    print("Deleting a single occurrence of 70 (had 3):")
    root = delete_node(root, 70)
    print("Inorder after deleting 70:")
    inorder(root)
    print("\n\nDeleting 50 (had 2):")
    root = delete_node(root, 50)
    print("Inorder after deleting 50:")
    inorder(root)
    print("\n\nDeleting 50 again (to completely remove it, it will be replaced by predecessor 40):")
    root = delete_node(root, 50)
    print("Inorder after fully deleting 50:")
    inorder(root)
    print("\n")
```

# Complexity Analysis

Here is the asymptotic complexity analysis for the modified methods. Let $n$ be the number of unique nodes in the BST, and $h$ be the height of the BST. We treat the `count` attribute processing as a small constant per element. 

## 1. `inorder(root)`, `preorder(root)`, `postorder(root)`
*   **T(n)**: The function recursively calls itself for the left and right subtrees and prints the node `count` times. $T(n) = T(n_{left}) + T(n_{right}) + O(1)$. Every node is visited exactly once.
*   **Big-O**: $O(n)$ - In the worst-case scenario, we visit all $n$ nodes.
*   **Big-Omega ($\Omega$)**: $\Omega(n)$ - Even in the best-case scenario, all nodes must be visited to traverse the tree.
*   **Big-Theta ($\Theta$)**: $\Theta(n)$ - Because the best and worst cases both grow linearly with $n$.

## 2. `print_level_order(root)`
*   **T(n)**: It calls `height(root)` taking $O(n)$ time, then iterates from $i=1$ to $h$, calling `print_specific_level(root, i)` which takes $O(n)$ worst-case time per level. Sum of nodes visited across all levels for a tree of height $h$ can be up to $O(n^2)$ if the tree is heavily unbalanced.
*   **Big-O**: $O(n^2)$ - In a skewed tree, level 1 takes 1 step, level 2 takes 2 steps, ..., level $n$ takes $n$ steps. The sum is $\approx n^2/2$.
*   **Big-Omega ($\Omega$)**: $\Omega(n \log n)$ - In a perfectly balanced BST, $h = \log n$. Finding elements level by level taking $O(n \log n)$ aggregate operations.
*   **Big-Theta ($\Theta$)**: There is no single tight bound covering all tree shapes, it varies from $n \log n$ to $n^2$.

## 3. `delete_node(root, key)`
*   **T(n)**: The method traverses down the tree to find the node, which takes $O(h)$ time, where $h$ is the height. If the node has two children, finding the inorder predecessor takes an additional $O(h)$ time. Total time is bounded by the height of the tree.
*   **Big-O**: $O(n)$ - If the tree is highly skewed (basically a linked list), the height is $n$.
*   **Big-Omega ($\Omega$)**: $\Omega(1)$ - In the best case, the root node is the one we want to delete, it only has one copy, and has 0 or 1 child, taking constant time. For traversing a balanced tree to find a node it would be $\Omega(\log n)$.
*   **Big-Theta ($\Theta$)**: $\Theta(h)$ which translates to $\Theta(\log n)$ for balanced trees and $\Theta(n)$ for skewed trees.

## 4. `print_specific_level(root, level)`
*   **T(n)**: The function traverses down to the specific level. In the worst case, it traverses all nodes up to that level.
*   **Big-O**: $O(n)$ - In the worst case (e.g., asking for the last level of a skewed tree), it visits $n$ nodes.
*   **Big-Omega ($\Omega$)**: $\Omega(1)$ - If the level is 1, it only visits the root and returns.
*   **Big-Theta ($\Theta$)**: Varies depending on the level requested and the tree's structure.

## 5. `print_leaf_nodes(root)` & `print_non_leaf_node(root)`
*   **T(n)**: The functions visit every node in the tree exactly once to check whether it's a leaf or non-leaf node. $T(n) = T(n_{left}) + T(n_{right}) + c$.
*   **Big-O**: $O(n)$ - All nodes are visited.
*   **Big-Omega ($\Omega$)**: $\Omega(n)$ - Even in the best case, you cannot skip nodes.
*   **Big-Theta ($\Theta$)**: $\Theta(n)$ - The time complexity grows strictly linearly with the number of nodes.
